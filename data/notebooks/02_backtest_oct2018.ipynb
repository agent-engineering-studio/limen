# Limen — backtest, October 2018 Southern-Italy storm

This notebook drives `limen backtest` (or its programmatic equivalent
`limen.cli.backtest.run`) over the Oct 28 — Nov 2 2018 window for
Puglia + Basilicata, then plots the §2.5 metrics (hit rate / FAR /
mean lead time) against the targets defined in
`regional_thresholds.yaml`.

**Requires** the DB to be seeded (`limen seed`), the IFFI data
ingested (`limen idrogeo-sync`, future prompt), and an Open-Meteo
connection (the archive API is free).

```python
import asyncio, os
from datetime import datetime, UTC
from limen.cli.backtest import run as run_backtest

os.environ['LIMEN_BACKTEST_START'] = datetime(2018, 10, 28, tzinfo=UTC).isoformat()
os.environ['LIMEN_BACKTEST_END']   = datetime(2018, 11,  2, tzinfo=UTC).isoformat()
asyncio.run(run_backtest())
```

## Reading the report

Reports land under `./reports/backtest_<aoi>_<start>_<end>.md`.
Look for **PASS / FAIL** annotations on hit rate, FAR, and mean lead
time. Tweak weights or component params in
`src/limen/config/regional_thresholds.yaml` and re-run.
